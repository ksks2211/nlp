```python
DATA_IN_PATH ="./data_in/"

file_list=['labeledTrainData.tsv.zip','unlabeledTrainData.tsv.zip','testData.tsv.zip']

for file in file_list:
    zipRef = zipfile.ZipFile(DATA_IN_PATH+file,'r')
    zipRef.extractall(DATA_IN_PATH)
    zipRef.close()
```

```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### tsv : tap으로 구분된 자료

```python
train_data = pd.read_csv(DATA_IN_PATH+"labeledTrainData.tsv",header=0,delimiter='\t',quoting=3)
```

```python
train_data.head()
```

# 데이터 크기

```python
# filesize
for file in os.listdir(DATA_IN_PATH):
    if 'tsv' in file and 'zip' not in file:
        print(file.ljust(30)+str(round(os.path.getsize(DATA_IN_PATH+file)/1000000,2))+'MB')
```

```python
# train data count
print('train data count : {}'.format(len(train_data)))
```

```python
# review len
train_length = train_data['review'].apply(len)
train_length.head()
```

```python
# visualize review len
plt.figure(figsize=(12,5))
plt.hist(train_length,bins=200,alpha=0.5,color='r',label='word')
plt.yscale('log',nonpositive='clip')
plt.axvline(np.mean(train_length),color='r',linestyle='dashed',linewidth=2,label="Mean")
plt.title('Log-Histogram of length of review')
plt.xlabel('Length of review')
plt.ylabel('Number of review')
plt.legend()
plt.show()
```

```python
print("Max len : {}".format(np.max(train_length)))
print("Min len : {}".format(np.min(train_length)))
print("Mean len : {}".format(np.mean(train_length)))
print("Standard deviation : {}".format(np.std(train_length)))
print("Median len : {}".format(np.median(train_length)))
print("1/4 : {}".format(np.percentile(train_length,25)))
print("3/4 : {}".format(np.percentile(train_length,75)))
```

```python
# boxplot
plt.figure(figsize=(12,5))
plt.boxplot(train_length,labels=['counts'],showmeans=True)
plt.show()
```

```python
# wordcloud
from wordcloud import WordCloud
cloud = WordCloud(width=800,height=600).generate(" ".join(train_data['review']))
plt.figure(figsize=(20,15))
plt.imshow(cloud)
plt.axis('off')
plt.show()
```

```python
# label proportion
fig, axe = plt.subplots(ncols=1)
fig.set_size_inches(6,3)
sns.countplot(x='sentiment',data=train_data, ax=axe,palette='Set2')
plt.show()
```

```python
print("positive review : {}".format(train_data['sentiment'].value_counts()[1]))
print("negative review : {}".format(train_data['sentiment'].value_counts()[0]))
```

```python
# word count 
train_word_counts = train_data['review'].apply(lambda x : len(x.split(' ')))
plt.figure(figsize=(12,5))
plt.hist(train_word_counts,bins=50,facecolor='r',label='train',alpha=0.5)
plt.title('Log-Histogram of word count in review',fontsize=15)
plt.axvline(np.mean(train_word_counts),color='r',linestyle='dashed',linewidth=2,label="Mean Word Counts")
plt.yscale('log',nonpositive='clip')
plt.legend()
plt.xlabel('Number of words',fontsize=15)
plt.ylabel('Number of reviews',fontsize=15)
plt.show()
```

```python
print("Max len : {}".format(np.max(train_word_counts)))
print("Min len : {}".format(np.min(train_word_counts)))
print("Mean len : {}".format(np.mean(train_word_counts)))
print("Standard deviation : {}".format(np.std(train_word_counts)))
print("Median len : {}".format(np.median(train_word_counts)))

print("1/4 : {}".format(np.percentile(train_word_counts,25)))
print("3/4 : {}".format(np.percentile(train_word_counts,75)))
```

```python
qmarks= np.mean(train_data['review'].apply(lambda x: '?' in x))
fullstop= np.mean(train_data['review'].apply(lambda x: '.' in x))
capital_first= np.mean(train_data['review'].apply(lambda x: x[0].isupper()))
capitals = np.mean(train_data['review'].apply(lambda x: max([y.isupper() for y in x])))
numbers= np.mean(train_data['review'].apply(lambda x: max([y.isdigit() for y in x])))

print("qmark : {:.2f}%".format(qmarks*100))
print("fullstop : {:.2f}%".format(fullstop*100))
print("capital_first : {:.2f}%".format(capital_first*100))
print("capital : {:.2f}%".format(capitals*100))
print("numbers : {:.2f}%".format(numbers*100))
```
